```python
import seaborn as sns
from matplotlib import pyplot as plt 
import numpy as np
import pandas as pd
import statistics as stat
from scipy.stats import iqr, shapiro, normaltest
```

```python
mean = 88
std_dev = 31
num_points = 23

# Generate normally distributed data
at_20_data = abs(np.random.normal(loc=mean, scale=std_dev, size=num_points))
is_20_data = abs(np.random.normal(loc=mean+30, scale=std_dev-5, size=num_points))

dataframe = {"p_at20": at_20_data,
             "p_is20": is_20_data
            }

df = pd.DataFrame(dataframe)
df["p_is20"].hist()
```
